```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from math import sin, exp
from scipy import optimize
```

```python
def f(x):
  return sin(x / 5.0) * exp(x / 10.0) + 5.0 * exp(-x / 2.0)  #от 1 до 30 включительно
```

```python
x0 = np.array(2)
optimize.minimize(f, x0, method='BFGS', bounds=(1, 30))
```

```python
x0 = np.array(30)
optimize.minimize(f, x0, method='BFGS', bounds=(1, 30))
```

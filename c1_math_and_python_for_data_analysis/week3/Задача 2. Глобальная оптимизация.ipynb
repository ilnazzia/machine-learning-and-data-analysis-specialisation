```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from math import sin, exp
from scipy import optimize 
```

```python
def f(x):
  return sin(x / 5.0) * exp(x / 10.0) + 5.0 * exp(-x / 2.0)  #от 1 до 30 включительно
```

```python
optimize.differential_evolution(f, [(1, 30)])
```

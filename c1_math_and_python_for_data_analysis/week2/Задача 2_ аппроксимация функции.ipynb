```python
%matplotlib inline
from scipy import linalg
import numpy as np
import matplotlib.pyplot as plt
from math import *
```

```python
def f(x):
  return sin(x / 5.0) * exp(x / 10.0) + 5.0 * exp(-x / 2.0)  #от 1 до 15 включительно
```

```python
f_v = np.vectorize(f)
x_base = np.arange(1, 15, 0.2)
y_base = f_v(x_base)
```

```python
def f_1(x, koeff):
  return koeff[0] + koeff[1] * x
```

```python
a = np.array([[1, 1], [1, 15]])
b = np.array([f(1), f(15)])
x = linalg.solve(a, b)
x
```

```python
y_1 = [f_1(1, x), f_1(15, x)]
plt.plot([1, 15], y_1, x_base, y_base)
```

```python
def f_2(x, koeff):
  return koeff[0] + koeff[1] * x + koeff[2] * x**2
```

```python
a = np.array([[1, 1, 1], [1, 8, 64], [1, 15, 225]])
b = np.array([f(1), f(8), f(15)])
x = linalg.solve(a, b)
x
```

```python
y_2 = [f_2(absc, x) for absc in x_base]
plt.plot(x_base, y_2, x_base, y_base)
```

```python
def f_3(x, koeff):
  return koeff[0] + koeff[1] * x + koeff[2] * x**2 + koeff[3] * x**3
```

```python
a = np.array([[1, 1, 1, 1], [1, 4, 16, 64], [1, 10, 100, 1000], [1, 15, 225, 3375]])
b = np.array([f(1), f(4), f(10), f(15)])
x = linalg.solve(a, b)
x
```

```python
y_3 = [f_3(absc, x) for absc in x_base]
plt.plot(x_base, y_3, x_base, y_base)
```

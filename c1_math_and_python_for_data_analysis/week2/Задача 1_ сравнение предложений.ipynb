```python
import re
import numpy as np
from scipy.spatial import distance
import pandas as pd
```

```python
sentences_file = open('/content/drive/MyDrive/Курсы/Специализация машинное обучение и анализ данных/Математика и Python для анализа данных/Неделя 2/sentences.txt', 'r')
```

```python
sentences_list = sentences_file.readlines()
```

```python
i = 0
for sentence in sentences_list:
  sentences_list[i] = sentence.lower()
  i += 1
```

```python
tokenized_sentences_list = []
for sentence in sentences_list:
  tokenized_sentence = re.split('[^a-z]', sentence)
  while '' in tokenized_sentence:
      tokenized_sentence.remove('')
  tokenized_sentences_list.append(tokenized_sentence)
```

```python
word_list = []
for tokenized_sentence in tokenized_sentences_list:
  word_list.extend(tokenized_sentence)
word_list_unique = list(set(word_list))
```

```python
NxD = []
for i in range(22):
  s = []
  for j in range(254):
    s.append(tokenized_sentences_list[i].count(word_list_unique[j]))
  NxD.append(s)
```

```python
df = pd.DataFrame(NxD)
df
```

```python
distances = []
for i in range(1, 22):
  distances.append(distance.cosine(NxD[0], NxD[i]))
distances
```
